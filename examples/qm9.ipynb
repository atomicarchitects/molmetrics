```python
print(Bond(Atom('C'), Atom('C'), 1.0))
```

```python
bond_lengths = mols.bond_lengths()
for bond, lengths in bond_lengths.items():
    plt.hist(lengths, bins=20)
    plt.title(f"{bond[0].symbol}-{bond[1].symbol} Bond Lengths")
    plt.show()
```
